```python
# Automatically extract the time slot columns (assuming they start from a certain column index)
time_slot_columns = df.columns[10:]  
```

```python
time_slot_columns
```

```python
# Use the dynamically extracted time slot columns in the melt function
df_melted = pd.melt(df, 
                    id_vars=['SCATS Number', 'Location', 'CD_MELWAY', 'NB_LATITUDE', 'NB_LONGITUDE', 
                             'HF VicRoads Internal', 'VR Internal Stat', 'VR Internal Loc', 
                             'NB_TYPE_SURVEY', 'Date'], 
                    value_vars=time_slot_columns, 
                    var_name='Start Time', 
                    value_name='Vehicle Flow')
```

```python
df_melted.head(10)
```

```python
df_melted['Date'] = pd.to_datetime(df_melted['Date'])
```

```python
# Set Date as the index
df_melted.set_index('Date', inplace=True)
```

```python
# Sort by Date index
df_melted.sort_index(inplace=True)
```

```python
df_melted.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le_location = LabelEncoder()
le_melway = LabelEncoder()

# Apply LabelEncoder to Location and CD_MELWAY
df_melted['Location'] = le_location.fit_transform(df_melted['Location'])
df_melted['CD_MELWAY'] = le_melway.fit_transform(df_melted['CD_MELWAY'])
```

```python
df_melted.head(10)
```

```python
df_melted.to_csv('data/Boroondara.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df_melted, test_size=0.2, shuffle=False)
```

```python
# Save the training set
train_df.to_csv('data/train.csv')
```

```python
# Save the test set
test_df.to_csv('data/test.csv')
```
